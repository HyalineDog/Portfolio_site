```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Issue': ['Data accessibility and clarity', 'Navigation issues', 'Data visualization problems',
              'Lack of visual feedback', 'Poor color contrast', 'Onboarding process', 'Information presentation'],
    'Frequency': [5, 4, 3, 2, 2, 1, 1],
    'Severity': [4, 4, 3, 3, 2, 2, 2]
}

df = pd.DataFrame(data)
df = df.sort_values('Frequency', ascending=False)
df['Rank'] = range(1, len(df) + 1)

plt.figure(figsize=(12, 8))
colors = ['#ff0000', '#ff0000', '#ff9900', '#ff9900', '#ffff00', '#ffff00', '#ffff00']

for i in range(len(df)):
    plt.scatter(df['Rank'][i], df['Severity'][i], s=df['Frequency'][i]*200, color=colors[i], alpha=0.7)
    plt.text(df['Rank'][i]+0.1, df['Severity'][i], df['Issue'][i], fontsize=12)

plt.xlabel('Rank')
plt.ylabel('Severity')
plt.title('Usability Issues - Frequency and Severity')
plt.xlim(0, len(df)+1)
plt.ylim(0, 6)
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 8))
colors = ['#ff0000', '#ff0000', '#ff9900', '#ff9900', '#ffff00', '#ffff00', '#ffff00']

for i in range(len(df)):
    plt.scatter(df['Rank'][i], df['Severity'][i], s=df['Frequency'][i]*200, color=colors[i], alpha=0.7)
    plt.text(df['Rank'][i]+0.1, df['Severity'][i], df['Issue'][i], fontsize=12)

# Add legend
legend_labels = ['Data accessibility and clarity', 'Navigation issues', 'Data visualization problems',
                 'Lack of visual feedback', 'Poor color contrast', 'Onboarding process', 'Information presentation']
legend_handles = [plt.scatter([], [], s=200, color=colors[i], alpha=0.7) for i in range(len(df))]
plt.legend(legend_handles, legend_labels, loc='upper right')

plt.xlabel('Rank')
plt.ylabel('Severity')
plt.title('Usability Issues - Frequency and Severity')
plt.xlim(0, len(df)+1)
plt.ylim(0, 6)
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = {
    'Issue': ['Data accessibility and clarity', 'Navigation issues', 'Data visualization problems',
              'Lack of visual feedback', 'Poor color contrast', 'Onboarding process', 'Information presentation'],
    'Frequency': [5, 4, 3, 2, 2, 1, 1],
    'Severity': [4, 4, 3, 3, 2, 2, 2]
}

df = pd.DataFrame(data)

plt.figure(figsize=(12, 8))
colors = ['#ff0000', '#ff0000', '#ff9900', '#ff9900', '#ffff00', '#ffff00', '#ffff00']

for i in range(len(df)):
    x = [df['Severity'][i]] * df['Frequency'][i]
    y = np.random.normal(df['Frequency'][i], 0.1, size=df['Frequency'][i])
    plt.scatter(x, y, color=colors[i], alpha=0.7)
    plt.text(df['Severity'][i]+0.1, df['Frequency'][i], df['Issue'][i], fontsize=12)

plt.xlabel('Severity')
plt.ylabel('Frequency')
plt.title('Usability Issues - Frequency and Severity')
plt.xlim(0, 6)
plt.ylim(0, max(df['Frequency'])+1)
plt.grid(True)
plt.show()
```
